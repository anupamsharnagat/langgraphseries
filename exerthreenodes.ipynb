```python
#In this excersise lets try 3 nodes and maybe diffrent inputs
from typing import Dict,TypedDict
from langgraph.graph import StateGraph
```

```python
class myState(TypedDict):
    name: str
    age: str
    skills: list
    final: str

def first_node(state: myState) -> myState:
    state["final"] = f"Hello {state['name']}"
    return state

def second_node(state: myState) -> myState:
    state["final"] = state["final"] + f", your age is {state['age']}"
    return state

def third_node(state: myState) -> myState:
    state["final"] = state["final"] + f", your skills are {state['skills']}"
    return state

graph = StateGraph(myState) #creating a graph with state of type myState which we defined above
graph.add_node("first", first_node)
graph.set_entry_point("first") #setting first node as entry point of graph where execution will start when we invoke app created by compiling this graph
graph.add_node("second", second_node)
graph.add_node("third", third_node)

graph.add_edge("first", "second")
graph.add_edge("second", "third")

app = graph.compile()

#now we can invoke the app
result = app.invoke({
    "name": "John",
    "age": "30",
    "skills": ["Python", "JavaScript"]
})  

print(result["final"])

```
