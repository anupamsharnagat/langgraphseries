```python
from typing import Dict,TypedDict,List
from langgraph.graph import StateGraph #like a framework to design and manage flow of task
```

```python
class AgentState(TypedDict):
    values: List[int]
    name: str
    operation: str
    result: str
```

```python
def process(state: AgentState) -> AgentState:
    if state["operation"] == 'sum':
        state["result"] = str(sum(state["values"]))
    elif state["operation"] == 'multiply':
        result = 1
        for v in state["values"]:
            result *= v
        state["result"] = str(result)   
    else:
        state["result"] = "Invalid operation"
    state["result"] = f"{state['name']} performed {state['operation']} on {state['values']} and got result {state['result']}"
    return state

#graph = StateGraph[AgentState]() #creating a graph with state type as AgentState
graph = StateGraph(AgentState) #creating a graph with state type as AgentState
graph.add_node("process", process) #adding a node to graph with name "process" and function process which will take input state and return output state
graph.set_entry_point("process")
graph.set_finish_point("process")

app = graph.compile() #compiling graph to create an app which can be invoked with input to get output
result = app.invoke({"name": "Anupam", "operation": "multiply", "values": [1, 2, 3, 4]}) #invoking app with initial state having name as "Alice", operation as "sum" and values as [1, 2, 3] to get output state with modified result
print(result)
```
