```python
from typing import Dict,TypedDict
from langgraph.graph import StateGraph #like a framework to design and manage flow of task

#define state of graph
class MyState(TypedDict):
    message1: str

#now define nodes of graph
def first_node(state:MyState)->MyState:
    #this is first node of graph which takes state as input and returns state as output
    state["message1"] = "Hello " + state["message1"] #modifying message in state by adding "Hello " before it
    return state

#now we will create graph by providing nodes and initial state
graph = StateGraph(MyState)

graph.add_node("greeter",first_node) #adding first node to graph

# now start point of graph is first_node and end point is also first_node as we have only one node in graph
graph.set_entry_point("greeter")
graph.set_finish_point("greeter")

app = graph.compile() #compiling graph to create an app which can be invoked with input to get output

result = app.invoke({"message1": "Anupam Sharnagat"}) #invoking app with initial state having message1 as "Alice" to get output state with modified message1

result["message1"] #printing message1 from result state which is output of graph
```
